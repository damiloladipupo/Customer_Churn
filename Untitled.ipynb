#### DATA PREPROCESSING

```python
#separating the features and target variable 
X = mydf.drop('Churn', axis=1)
y = mydf['Churn']
```

```python
#splitting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=5)
```

```python
# import the Random Under Sampler object
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=5)
u_X_train, u_y_train= rus.fit_resample(X_train, y_train)
```

```python
u_y_train.dtype
```

```python
# import the Random over Sampler object
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=5)
o_X_train, o_y_train= rus.fit_resample(X_train, y_train)
```

```python
#checking classification accuracy
TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]
```

```python
# print classification accuracy

gnb_classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print('Classification accuracy : {0:0.6f}'.format(gnb_classification_accuracy))
```

```python
#printing classification error

gnb_classification_error = (FP + FN) / float(TP + TN + FP + FN)

print('Classification error : {0:0.6f}'.format(gnb_classification_error))
```

```python
# print precision score

gnb_precision = TP / float(TP + FP)


print('Precision : {0:0.6f}'.format(gnb_precision))
```

```python
gnb_recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.6f}'.format(gnb_recall))
```

```python
gnb_true_positive_rate = TP / float(TP + FN)


print('True Positive Rate : {0:0.6f}'.format(gnb_true_positive_rate))
```

```python
gnb_false_positive_rate = FP / float(FP + TN)


print('False Positive Rate : {0:0.6f}'.format(gnb_false_positive_rate))
```

```python
gnb_specificity = TN / (TN + FP)

print('Specificity : {0:0.6f}'.format(gnb_specificity))
```

```python
# compute ROC AUC
# ROC AUC is a single number summary of classifier performance. The higher the value, the better the classifier.

#ROC AUC of our model approaches towards 1. So, we can conclude that our classifier does a good job in predicting whether 
#there will be churn or not.


from sklearn.metrics import roc_auc_score

ROC_AUC = roc_auc_score(y_test, y_pred1)

print('ROC AUC : {:.4f}'.format(ROC_AUC))
```
