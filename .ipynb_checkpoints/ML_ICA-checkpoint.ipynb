```python
#rebuilding the classification model using different K values from 3 to 19
# Define a range of k values
k_values = list(range(3, 20))

# Create an empty dictionary to store k values and their corresponding accuracy scores
accuracy_scores = {}

# Iterate through the k values
for k in k_values:
    # Instantiate the model with the current k value
    knn_model = KNeighborsClassifier(n_neighbors=k)
    
    # Fit the model to the training set
    knn_model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = knn_model.predict(X_test)
    
    # Calculate the accuracy score and store it in the dictionary
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores[k] = accuracy

# Print the accuracy scores for each k value
for k, accuracy in accuracy_scores.items():
    print('K = {}: Accuracy = {:.7f}'.format(k, accuracy))
```

```python
#our model performed best at k = 6 with 0.7785664 accuracy. however, to avoid ties, we will be using an odd value of k.
#from the above, k = 5, 16 and 18 has the next hiher accuracy of 0.7764372. Hence, we will use k = 5
```
